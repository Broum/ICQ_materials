# Rotační brány



Každá jednoqubitová unitární operace odpovídá **rotaci Blochovy sféry** kolem nějaké osy. Pro úplný popis libovolné jednoqubitové brány stačí tři rotační úhly (Eulerovy úhly). Základní rotační brány jsou:

- $R_x(\theta)$ – rotace kolem osy **X**
- $R_y(\theta)$ – rotace kolem osy **Y**
- $R_z(\theta)$ – rotace kolem osy **Z**

## Definice rotačních bran přes exponenty Pauliho matic
Rotační brány definujeme jako exponenciály Pauliho matic:

$$
R_x(\theta) = e^{-i\frac{\theta}{2} X}, \quad
R_y(\theta) = e^{-i\frac{\theta}{2} Y}, \quad
R_z(\theta) = e^{-i\frac{\theta}{2} Z}.
$$

Pauliho matice:
$$
X = \begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix},\quad
Y = \begin{bmatrix} 0 & -i \\ i & 0 \end{bmatrix},\quad
Z = \begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix}.
$$

## Explicitní maticové tvary
Rozepíšeme exponenciály do uzavřených tvarů:

$$
R_x(\theta) = \begin{bmatrix}
\cos\tfrac{\theta}{2} & -i\sin\tfrac{\theta}{2} \\
-i\sin\tfrac{\theta}{2} & \cos\tfrac{\theta}{2}
\end{bmatrix},\quad
R_y(\theta) = \begin{bmatrix}
\cos\tfrac{\theta}{2} & -\sin\tfrac{\theta}{2} \\
\sin\tfrac{\theta}{2} & \cos\tfrac{\theta}{2}
\end{bmatrix},\quad
R_z(\theta) = \begin{bmatrix}
e^{-i\theta/2} & 0 \\
0 & e^{i\theta/2}
\end{bmatrix}.
$$

## Důležité speciální případy
Následující brány jsou jen **rotace o speciální úhly**:

| Rotace | Úhel $\theta$ | Výsledek | Běžný název |
|---|---|---|---|
| $R_x(\pi)$ | 180° | $= X$ | Pauli-X |
| $R_y(\pi)$ | 180° | $= Y$ | Pauli-Y |
| $R_z(\pi)$ | 180° | $= Z$ | Pauli-Z |
| $R_z(\pi/2)$ | 90° | $= S$ | fázová brána S |
| $R_z(\pi/4)$ | 45° | $= T$ | fázová brána T |


## Geometrická interpretace na Blochově sféře
- $R_x(\theta)$: rotace kolem osy X (východ–západ)
- $R_y(\theta)$: rotace kolem osy Y (sever–jih)
- $R_z(\theta)$: rotace kolem osy Z (změna fáze – pohyb po rovníku)

# Deutschův algoritmus

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

def deutsch_function(case: int):
    # This function generates a quantum circuit for one of the 4 functions
    # from one bit to one bit
 
    if case not in [1, 2, 3, 4]:
        raise ValueError("`case` must be 1, 2, 3, or 4.")
 
    f = QuantumCircuit(2)
    if case in [2, 3]:
        f.cx(0, 1)
    if case in [3, 4]:
        f.x(1)
    return f

    
display(deutsch_function(2).draw(output="mpl"))
```

```python
qc = QuantumCircuit(2,1)
qc.x(1)
# qc.x(0) ### vyzkoušejte, co se stane, když dám i negaci na q_0 nebo naopak negaci nedám ani na q_1
qc.h(0)
qc.h(1)

qc.barrier()

qc.compose(deutsch_function(1), inplace='True')

qc.barrier()

qc.h(0)

qc.measure(0,0)

qc.draw(output='mpl')

```

```python
result = AerSimulator().run(qc, shots=1, memory=True).result()
measurements = result.get_memory()
if measurements[0] == "0":
    print("constant")
else:
    print("balanced")
```

# Deutschův - Jozsův algoritmus
zobecnění Deutschova algoritmu pro Deutschův - Jozsův problém:

**Vstup:**  
Funkce  $ f : \{0,1\}^n \rightarrow \{0,1\} $

**Předpoklad:**  
$ f $ je buď konstantní nebo vyvážená. Jiné funkce nejsou povoleny a brány v potaz. 

**Výstup**  
$ \begin{cases}
0 & \text{pokud } f \text{ je konstantní} \\
1 & \text{pokud } f \text{ je vyvážená}
\end{cases}$

```python
# implementace obecné funkce pro D-J problém: https://quantum.cloud.ibm.com/learning/en/courses/fundamentals-of-quantum-algorithms/quantum-query-algorithms/deutsch-jozsa-algorithm


def dj_query(num_qubits):
    # Create a circuit implementing for a query gate for a random function
    # satisfying the promise for the Deutsch-Jozsa problem.
 
    qc = QuantumCircuit(num_qubits + 1)
 
    if np.random.randint(0, 2):
        # Flip output qubit with 50% chance
        qc.x(num_qubits)
    if np.random.randint(0, 2):
        # return constant circuit with 50% chance
        return qc
 
    # Choose half the possible input strings
    on_states = np.random.choice(
        range(2**num_qubits),  # numbers to sample from
        2**num_qubits // 2,  # number of samples
        replace=False,  # makes sure states are only sampled once
    )
    print(on_states)
    def add_cx(qc, bit_string):
        for qubit, bit in enumerate(reversed(bit_string)):
            if bit == "1":
                qc.x(qubit)
        return qc
 
    for state in on_states:
        qc.barrier()  # Barriers are added to help visualize how the functions are created.
        qc = add_cx(qc, f"{state:0b}")
        qc.mcx(list(range(num_qubits)), num_qubits)
        qc = add_cx(qc, f"{state:0b}")
 
    qc.barrier()
 
    return qc
```

```python
def compile_circuit(function: QuantumCircuit):
    # Compiles a circuit for use in the Deutsch-Jozsa algorithm.
 
    n = function.num_qubits - 1
    qc = QuantumCircuit(n + 1, n)
    qc.x(n)
    qc.h(range(n + 1))
    # qc.barrier()
    qc.compose(function, inplace=True)
    qc.barrier()
    qc.h(range(n))
    qc.barrier()
    qc.measure(range(n), range(n))
    return qc
```

```python
def dj_algorithm(function: QuantumCircuit):
    # Determine if a function is constant or balanced.
 
    qc = compile_circuit(function)
    
    result = AerSimulator().run(qc, shots=1, memory=True).result()
    measurements = result.get_memory()
    if "1" in measurements[0]:
        return "balanced"
    return "constant"
```

```python
f = dj_query(3)
# display(f.draw("mpl"))
full_qc = compile_circuit(f) 
display(dj_algorithm(f))
display(full_qc.draw(output="mpl"))
```

# Bernsteinův-Vaziraniho problém
binární skalární součin (binaryy dot product, nejsem si úplně jistý český názvem): 
$ x \cdot y = \begin{cases}
1 & \text{pokud } x_{n-1}y_{n-1} + \cdots + x_0 y_0  \text{ je liché číslo } \\
0 & \text{pokud } x_{n-1}y_{n-1} + \cdots + x_0 y_0  \text{ je sudé číslo }
\end{cases}$


**Vstup:**  
Funkce  $ f : \{0,1\}^n \rightarrow \{0,1\} $

**Předpoklad:**  
Existuje binární řetězec $ s = s_{n-1}\cdots s_0 $ pro který $ f(x) = s \cdot x  $ pro x $ \epsilon \Sigma^n $
 
**Výstup**  
řetezec $ s $

```python
def bv_query(s):
    # Create a quantum circuit implementing a query gate for the
    # Bernstein-Vazirani problem.
 
    qc = QuantumCircuit(len(s) + 1)
    for index, bit in enumerate(reversed(s)):
        if bit == "1":
            qc.cx(index, len(s))
    return qc
 
 
# display(bv_query("11001").draw(output="mpl"))
```

```python
def bv_algorithm(function: QuantumCircuit):
    qc = compile_circuit(function)
    result = AerSimulator().run(qc, shots=1, memory=True).result()
    return result.get_memory()[0]
 
full_qc = compile_circuit(bv_query("1011")) 
display(full_qc.draw(output="mpl"))
display(bv_algorithm(bv_query("1011")))
```

```python
# matrix_exponential_rotations.py
# Notebook-style script: explains matrix exponential and constructs Rx, Ry, Rz

from IPython.display import Markdown, display
import numpy as np
from scipy.linalg import expm
import sympy as sp

# Explanation
explanation = """
### Matrix exponential & rotation gates

For Pauli matrices the exponential collapses to cos/sin terms because sigma^2 = I.
The generator -i*(theta/2)*sigma gives the single-qubit rotation gates.
"""
display(Markdown(explanation))

# Pauli matrices
I = np.eye(2, dtype=complex)
sigma_x = np.array([[0, 1], [1, 0]], dtype=complex)
sigma_y = np.array([[0, -1j], [1j, 0]], dtype=complex)
sigma_z = np.array([[1, 0], [0, -1]], dtype=complex)

theta = 1.2345  # radians

Rx_from_expm = expm(-1j * theta/2 * sigma_x)
Ry_from_expm = expm(-1j * theta/2 * sigma_y)
Rz_from_expm = expm(-1j * theta/2 * sigma_z)

cos = np.cos(theta/2)
sin = np.sin(theta/2)

Rx_closed = np.array([[cos, -1j*sin], [-1j*sin, cos]], dtype=complex)
Ry_closed = np.array([[cos, -sin], [sin, cos]], dtype=complex)
Rz_closed = np.array([[np.exp(-1j*theta/2), 0], [0, np.exp(1j*theta/2)]], dtype=complex)

def matrix_equal(a, b, tol=1e-12):
    return np.allclose(a, b, atol=tol, rtol=0)

print(f"Theta = {theta} rad")
print("Rx equal:", matrix_equal(Rx_from_expm, Rx_closed))
print("Ry equal:", matrix_equal(Ry_from_expm, Ry_closed))
print("Rz equal:", matrix_equal(Rz_from_expm, Rz_closed))

# Symbolic check
theta_sym = sp.symbols('theta', real=True)
sx_sym = sp.Matrix([[0, 1], [1, 0]])
M = -sp.I * theta_sym/2 * sx_sym
Mexp = sp.exp(M)
Mexp_simpl = sp.simplify(Mexp)

print("\nSymbolic exp(-i theta/2 sigma_x) =")
sp.pprint(Mexp_simpl, use_unicode=True)
```
