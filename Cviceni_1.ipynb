````python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector, plot_histogram
from numpy import pi

qc = QuantumCircuit(3)

# ---------- 1-qubit gates -------------

# qc.x(0)

qc.h(0)
qc.h(1)
qc.h(2)

# qc.z(0)

# qc.x(0)
# qc.h(1)
# qc.s(0)

# qc.t(0)
# qc.y(0)
# # qc.t(1)

#qc.sx(0)

# qc.p(pi/8,0)

# ------------ 2-qubit gates -------------


# qc.cx(0,1)

#qc.cz(0,1)

# qc.cp(pi/8,0,1)

#qc.swap(0,1)

# qc.t(1)

# qc.draw(output='mpl') # nicer figire
qc.draw()

````

````python
state = Statevector(qc)
print(state) 
````

````python
plot_bloch_multivector(state)
````

````python
qc = QuantumCircuit(2)

#qc.x(0)
#qc.x(1)

qc.h(0)
# qc.x(1)
qc.cx(0,1)

# qc.z(1)



qc.draw(output='mpl')
````

````python
state = Statevector(qc)
print(state)
````

````python
plot_bloch_multivector(state)
````

````python
qc.measure_all()

qc.draw(output='mpl')
````

````python
from qiskit_aer import AerSimulator
backend = AerSimulator()
results = backend.run(qc, shots=10000).result()
counts = results.get_counts()
plot_histogram(counts)
````
